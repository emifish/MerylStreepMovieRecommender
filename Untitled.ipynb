```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import re


# Step 1: Load data (use your actual cleaned data)
df = pd.read_csv("meryl_streep_movies.csv")

# Step 2: Prepare simplified features (genre + mood + numeric)
df['Genre'] = df['Genre'].fillna('').str.split('/')
df['Mood_Tags'] = df['Mood_Tags'].fillna('').str.split(', ')

from sklearn.preprocessing import MultiLabelBinarizer

mlb_genre = MultiLabelBinarizer()
mlb_mood = MultiLabelBinarizer()

df_genre = pd.DataFrame(mlb_genre.fit_transform(df['Genre']), columns=mlb_genre.classes_)
df_mood = pd.DataFrame(mlb_mood.fit_transform(df['Mood_Tags']), columns=mlb_mood.classes_)

# Numeric features
df['IMDb_Rating'] = pd.to_numeric(df['IMDb_Rating'], errors='coerce').fillna(0)
df['Num_Nominations'] = df['Awards'].str.count('Nomination', flags=re.IGNORECASE).fillna(0)
df['Num_Wins'] = df['Awards'].str.count('Win', flags=re.IGNORECASE).fillna(0)
numeric = df[['IMDb_Rating', 'Num_Nominations', 'Num_Wins']]

# Combine and scale
combined = pd.concat([df_genre, df_mood, numeric], axis=1)
scaler = StandardScaler()
features = scaler.fit_transform(combined)

# Step 3: Cluster
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(features)

# Step 4: Reduce to 2D with PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(features)

# Step 5: Plot
plt.figure(figsize=(10, 7))
scatter = plt.scatter(reduced[:, 0], reduced[:, 1], c=df['Cluster'], cmap='viridis', s=100, alpha=0.7)
plt.title("Meryl Streep Movie Clusters (via PCA)", fontsize=16)
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")

# Optional: Annotate with movie titles
for i, title in enumerate(df['Title']):
    plt.annotate(title, (reduced[i, 0], reduced[i, 1]), fontsize=8, alpha=0.7)

plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```
